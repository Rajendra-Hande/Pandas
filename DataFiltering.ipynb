## Filtering Data from DataFrame

```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset/PersonalInfo.csv')
df.head()
```

```python
df.Age > 20 # generate Pandas Series of boolean value
```

```python
df[df.Age > 20] # Filtering DataFrame based on the condition
```

```python
df[df.Age > 20]['Name']  # Series of Names whose Agr > 20
```

```python
df[df.Age > 20][['Name','Gender','Age']].head()
```

## Multiple Filtering Criteria

```python
df.head()
```

```python
df[(df.Gender == 'Male') & (df.Age > 20)] # Filtering Data Gender and Age wise
```

```python
df[(df.Gender == 'Male') | (df.Age > 20)] 
```

```python
df[df.City.isin(['Pune','Hyderabad'])] # df[(df.City=='Pune') | (df.City=='Hyderabad')]
```

```python
df[(df.Age > 20) & (df.City.isin(['Pune','Hyderabad']))]
```

```python
df[(df.Age > 20) & (df.City.isin(['Pune','Hyderabad']))]['Name']
```
