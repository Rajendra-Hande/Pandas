```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset/PersonalInfo.csv')
df.head()
```

```python
pd.get_dummies(df.Gender, prefix = 'Sex').head()
```

```python
Sex =  pd.get_dummies(df.Gender, prefix = 'Sex').iloc[:,1].head()
```

```python
df = pd.concat([df,Sex], axis = 1)
df.head()
```

```python
df.drop('Gender', axis = 1, inplace = True)
df.head()
```
